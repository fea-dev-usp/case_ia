# **1 Apresentação**

## **1.1 Contexto e motivação**

Em um cenário real, é muito comum termos a seguinte situação:

* **Muitos dados disponíveis**: bases grandes, com várias variáveis (idade, renda, uso de app, exames, histórico de compras etc.).

* **Poucos rótulos prontos**: quase nunca alguém marcou, linha a linha, quem é “cliente bom”, quem é “alto risco”, qual é o “perfil” do paciente, e assim por diante.

Alguns exemplos de dados não rotulados:

* Usuários de um app
  * Quantidade de logins, tempo de sessão, telas visitadas, funcionalidades
usadas…
  * Mas não há uma classificação prévia como “usuário iniciante”, “engajado”, “quase desistindo”.

* Pacientes com vários exames
  * Diversos resultados de exames, histórico de consultas, queixas principais…
  * Sem uma divisão pronta em “grupos de pacientes” com perfis parecidos.


Se os dados já possuem rótulos o ideal seria o aprendizado supervisionado, mas como visto anteriormente, no mundo real existem grandes chances da base de dados não ser rotulada. Diante disso, surge a pergunta central que define o papel de modelos de aprendizagem não supervisionado:

> O que dá para aprender desses dados sem gabarito?


Ou seja, como **descobrir estruturas, padrões e grupos** apenas olhando para as variáveis que temos, sem uma resposta correta informada de antemão?

É exatamente esse tipo de situação que o **Aprendizado Não Supervisionado (UL - Unsupervised Learning)** busca resolver, encontrando informações e fazendo combinações dos dados para extrair insights de como tais dados podem ser agrupados, como eles se organizam, sem essa resposta prévia.

Junto ao UL, iremos relacionar com a Análise exploratória. A **finalidade da Análise Exploratória de Dados (AED)** é examinar os dados previamente à aplicação de qualquer técnica estatística. Desta forma o analista consegue um entendimento básico de seus dados e das relações existentes entre as variáveis analisadas. Após a coleta e a digitação de dados em um banco de dados apropriado, o próximo passo é a análise descritiva. Esta etapa é fundamental, pois uma análise descritiva detalhada permite ao pesquisador familiarizar-se com os dados, organizá-los e sintetizá-los de forma a obter as informações necessárias do conjunto de dados para responder às questões que estão sendo estudadas.


## **1.2 Definição de aprendizado não supervisionado**

Para definirmos um conceito de aprendizado não supervisionado, podemos pensar primeiramente de uma forma mais intuitiva: Um modelo de aprendizado não supervisionado, é aquele que recebe apenas os dados de entrada(sem classificações) e tenta extrair padrões e regularidades por conta própria(não supervisionado), propondo formas de organizar esses dados.

Pensando em um exemplo, imagine que a sua mãe manda você lavar a louça e organizar os utensílios no escorredor para secar, não dizendo de que maneira ou como separar cada coisa. Mesmo sem saber essa maneira você organizou colocando as panelas com as panelas, os talheres com os talheres e os copos com os copos, ou seja, você virou um modelo de aprendizagem não supervisionado(apesar de nesse caso sua mãe estar de olho), colocando os itens que possuem mais semelhanças juntos, mesmo sem estar escrito que a panela é uma panela ou o se a caneca é um copo!

Partindo desse raciocínio, fica mais fácil de entender a definição formal:  Aprendizado Não Supervisionado é o conjunto de técnicas que, usando somente as **variáveis de entrada**, busca estruturas ocultas nos dados, como clusters (agrupamentos), componentes principais ou padrões de coocorrência , **sem** utilizar rótulos de saída.

Pontos importantes dessa definição:

* Não há uma variável objetivo (“target”) pré-definida.(não existe uma coluna alvo que o modelo deve adivinhar)
* O objetivo não é “acertar um rótulo”, mas revelar como os dados se organizam.
* A avaliação do resultado envolve interpretação e conhecimento de domínio, não só métricas numéricas.( o analista deve olhar para os clusters, por exemplo, dar nome aos perfis criados, e principalmente, observar se a classificação faz sentido com a realidade)

Alguns exemplos de uso de aprendizado não supervisionado são:

**Análise exploratória de dados**: Utilizado para descobrir padrões dentro da estrutura do conjunto de dados, obtendo análises profundas sobre tais informações.

**Segmentação de imagens**: Recorrente em áreas médicas, principalmente em radiologia, onde o modelo cria máscaras de áreas específicas (como ossos, ou órgãos) para análise futura. Não é necessário um modelo supervisionado pois, para a segmentação, é necessário apenas as características naturais da imagem.

**Sistemas de recomendação**: Como a estratégia não supervisionada é muito boa para procurar relações entre dados, é extremamente útil ao criar algoritmos de recomendações em redes sociais ou apps de streaming, por exemplo. Desse modo, o modelo consegue assimilar usuários de gostos parecidos e recomendar conteúdos parecidos entre eles.

**Processamento de linguagem natural**: De maneira similar ao sistema de recomendação, o modelo consegue achar padrões entre textos que não são tão perceptíveis em uma análise humana. Nesse mesmo aspecto, consegue achar, também, correlações entre partes específicas de textos, útil para análise de sentimento, por exemplo.

## **1.3 Comparação rápida: Supervisionado x Não supervisionado**


Para treinar modelos **supervisionados**, é necessário um conjunto de dados com os resultados que queremos (rótulos).Ou seja, para cada exemplo que temos, é obrigatório saber qual a sua saída correta. Dessa maneira, o modelo aprende ao comparar o seu resultado com o resultado real.

Já nos modelos **não supervisionados**, não é preciso os resultados esperados pré definidos. O conjunto de dados contém apenas as entradas, com o objetivo do modelo de descobrir padrões ou agrupar tais informações.

Portanto, temos as seguintes diferenças de características entre os modelos:

1. Rotulagem do conjunto de dados
   - Supervisionado: Deve-se aplicar os resultados reais de cada exemplo presente na coleção.
   - Não Supervisionado: Consegue trabalhar sem possuir uma rotulagem pré-definida.
2. Objetivo do modelo
   - Supervisionado: Prever valores de dados baseados no seu histórico ou classificá-los baseado em classes pré-fixadas.
   - Não Supervisionado: Analisar e explorar novos dados para achar padrões na estrutura do conjunto de dados.
3. Tipos de algoritmos
   - Supervisionado: Regressão e classificação
   - Não Supervisionado: Agrupamento de dados
4. Trade-offs:
   - Supervisionado: Produz resultados muito acurados, porém é necessário um maior trabalho humano em sua preparação.
   - Não Supervisionado: Pode haver alguns erros e resultados não assertivos, contudo, não é necessário um grande trabalho manual para rodá-lo.


## **1.4 Quando UL é útil (e quando não é)**

O motivo mais óbvio para usar aprendizado não supervisionado é simplesmente quando os dados disponíveis não estão rotulados, ou existem dados rotulados, mas em menor quantidade. Uma das maiores vantagens do aprendizado não supervisionado é que costuma ser muito mais fácil obter dados não rotulados do que dados rotulados. Por exemplo, tratando-se de modelos que trabalham com linguagem, possuímos uma quantidade gigantesca de dados não rotulados na internet, seria impossível obter mesmo uma fração da quantidade de dados se nos restringirmos a dados rotulados.

Geralmente se possuímos dados rotulados é preferível usar aprendizado supervisionado, pois costuma ser mais preciso, porém aprendizado não supervisionado ainda pode apresentar vantagens. O objetivo do aprendizado não supervisionado é encontrar padrões e estruturas dentro de um conjunto de dados, essa capacidade pode ser usada de inúmeras maneiras. Dados nem sempre estão estruturados da melhor forma para uso, podemos ter features demais, features com alta correlação, features pouco relevantes, relações complexas e outros problemas que podem apresentar um desafio a muitos modelos. Modelos de aprendizado não supervisionado podem transformar os dados, encontrar relações e simplificar as features de maneira a tentar extrair as informações que importam de um conjunto de dados complexo, facilitando a análise e potencialmente até passando o resultado para um algoritmo supervisionado.

Por fim, um caso em que aprendizado não supervisionado tem maior dominância é em tarefas generativas. Modelos generativos focam em encontrar relações entre os dados e como o nome sugere, podem ser usados para gerar novos dados, por exemplo, em geração de imagens. Porém muitas tarefas podem ser resolvidas tanto com aprendizado não supervisionado quanto com aprendizado supervisionado e nem sempre é claro qual é melhor.

## **1.5 Análise exploratória de dados (AED)**


Antes de realizar qualquer modelagem não supervisionada, é essencial que tenhamos total entendimento do nosso conjunto de dados e como seus atributos estão distribuídos e relacionados entre si. Para isso, realizamos a análise exploratória de dados antes de quaisquer suposições.

É uma etapa extremamente importante para confirmar que seus dados utilizados são válidos para quaisquer objetivo do e resultados desejados, além de certificar que o contexto utilizado é o correto. Além disso o processo de AED possui as seguintes finalidades:

1.   Detecção de erros
2.   Verificação de suposições
3.   Determinar a relação entre os atributos e entre as variáveis de entrada e saída.

Podemos falar que quaisquer métodos que não envolvem modelagem estatística e inferência fazem parte da análise exploratória.


## **1.6 Tipos de análise exploratória**

Métodos de análise exploratória podem ser classificados de duas maneiras:

*   Análise univariada: Objetivo de analisar apenas uma variável do conjunto de dados.
*   Análise multivariada: Analisa a relação entre duas ou mais variáveis do conjunto de dados. Dizemos que a análise é bivariada quando analisamos apenas dois atributos apenas.

Para essas análises, podemos realizar duas estratégias:

*   Métodos gráficos: Para análises univariadas, são muito bons para entender detalhes da distribuição de cada atributo, como padrões na distribuição e outliers. Já na análise multivariada, buscam explorar a relação entre variáveis com a correlação sendo uma das métricas mais importantes.
*   Métodos não gráficos: Costumam envolver métricas que buscam sumarizar os dados, fornecendo informações numéricas sobre a distribuição dos dados, em análises univariadas

Portanto, temos os seguintes tipos de análise exploratória de dados:

1.   Univariada não gráfica
2.   Univariada gráfica
3.   Multivariada não gráfica
4.   Multivariada gráfica


# **2 Principais conceitos**


## **2.1 Tipos de variáveis**


Variáveis podem ser de dois tipos, categóricas/qualitativas ou numéricas/quantitativas. Variáveis categóricas não possuem um valor numérico, cada variável tem uma quantidade finita de valores que pode assumir (categorias) podendo ser ordinais (discorda, neutro, concorda; ruim, satisfatório, bom; etc) ou nominais (cidade, tipo sanguíneo, cor, etc).

Variáveis categóricas ordinais possuem uma ordem clara, porém não necessariamente existe uma relação numérica apropriada entre seus valores, pois continuam não sendo números.

Já variáveis categóricas nominais não possuem qualquer ordem de grandeza entre elas. Seja qual for o caso, é sempre necessário transformar seus valores em números para que os modelos possam compreendê-las. Um método bastante apropriado para isso é o One-Hot Encoding, que transforma as features de uma maneira que não cria falsas relações de grandeza.

Entre as métricas que mais nos interessam em uma variável categórica estão frequência, alcance e moda.

Já as variáveis quantitativas representam valores numéricos, podendo ser contínuos (números reais) ou discretos (números inteiros). As métricas e distribuições que variáveis numéricas podem assumir são muito mais variadas. Algumas características importantes são: média, mediana, alcance, moda, variância, formato da distribuição, skewness, kurtosis, outliers e outras.

Entender as variáveis é fundamental para decidir como técnicas de aprendizado não supervisionado podem melhor beneficiar seu conjunto de dados.


## **2.2 Dados, instâncias e features**


Já temos bastante familiaridade com os conceitos de dados, observações e features, o que muda agora é a forma como vamos encarar o problema. Ao se tratar de aprendizado não supervisionado não temos mais um target que queremos prever, ao invés disso todos os elementos podem ser alvo de análise.

A aplicação mais comum de aprendizado não supervisionado é encontrar relações entre as features, descobrindo padrões e estruturas, mas podemos também tentar encontrar outliers entre as observações, realizar observações gerais sobre o conjunto de dados ou transformar os dados.


## **2.3 Espaço de características**


Aqui é a mesma definição que foi feita no módulo inicial, que será essencial para o UL, pois podemos imaginar os dados geometricamente. Relembrando, cada feature ou característica representará uma variável que vira um eixo no espaço, e cada observação um ponto nesse espaço, por exemplo:

*   Eixo x representa cor do cabelo
*   Eixo y representa idade
*   Eixo z representa o gênero

Agora, cada pessoa será uma observação, sendo representada por um ponto nesse espaço de acordo com suas características. O espaço de características é onde todas as possíveis combinações de features podem existir.

Agora para entender a relação dessa definição com o UL, podemos pensar assim:
O dataset é uma grande nuvem de pontos em um espaço de características.
O UL vai tentar encontrar estruturas nessa nuvem:

    regiões densas (clusters), direções principais de variação (componentes), pontos isolados (anomalias), ...

Essa visão de “nuvem de pontos” é a base para entender:

*   Por que distância e similaridade importam tanto.
*   Por que a escala das variáveis pode distorcer o resultado (se um eixo é gigantesco e o outro pequeno, um domina a distância).


## **2.4 Similaridade e distância**

Quase todas as técnicas de UL, de um jeito ou de outro, respondem à pergunta:

> Quais pontos são **parecidos** e quais são muito **diferentes**?

Para transformar essa ideia intuitiva em algo que o algoritmo consegue usar, precisamos de um conceito de **distância**.

Antes de entrar nesse conceito, é importante entender a diferença entre “parecido” e “diferente”, que é bem simples: dois pontos são considerados parecidos quando têm valores semelhantes nas principais variáveis (ex: idade parecida, renda parecida, comportamento de compra semelhante, etc.), e são considerados diferentes quando têm valores muito distantes em uma ou mais variáveis relevantes. Agora usando o espaço de características, em termos geométricos, quantos mais próximos  dois pontos estão, mais similares eles são e quanto mais distantes mais diferentes.

Então pensando nessa definição, a distância vai dizer ao modelo o quão similares determinadas observações são, auxiliando-o a enxergar padrões e organizar os dados  Existem várias formas de medir “distância” entre dois pontos(euclidiana, similaridade do cosseno, distância Manhattan etc) mas a ideia central é sempre a mesma:

*   Quanto **menor** a distância, mais **similares** as observações.
*   Quanto **maior** a distância, mais **diferentes** elas são

Logo, a lógica é: o algoritmo calcula a distância e com base nisso tenta agrupar ou organizar os dados.

Nesse sentido, também é válido ressaltar a importância da padronização e normalização dos dados, pois se as variáveis têm escalas muito diferentes, a distância pode ficar **dominada** por uma delas, como já vimos nos módulos anteriores.


## **2.5 Cluster x Classe**

Imagine tentar entender uma música com várias notas de vários instrumentos. Ao escutar pela primeira vez, é difícil identificar de qual instrumento vem cada nota, esse conjunto de notas parecem um ruído dificilmente reconhecível. Porém, a cada reprodução da música, você vai percebendo e reconhecendo se cada nota vem do contrabaixo, trompete, piano, saxofone ou da bateria! Assim, após algumas escutas, você consegue identificar a linha de notas do contrabaixo e separá-las do trompete.

Esse processo que você realizou é a estratégia de agrupamento de dados, percebe que é muito mais simples analisar a música após separar o que vem de cada instrumento? Esse mesmo pensamento é aplicado para um conjunto de dados qualquer, onde é muito mais fácil analisá-los após o agrupamento. Para isso, temos duas estratégias de agrupamento: Clusterização e Classificação.

Os modelos de clusterização e de classificação são intrinsecamente diferentes em sua jornada. Portanto, é essencial saber a estrutura do seu projeto, do seu conjunto de dados e, principalmente, a diferença entre os dois modelos para escolher o mais apropriado para seu projeto.

### **3.5.1 Classificação**

Agora, imagine suas playlists de música, cada uma para um gênero de música específico: jazz, samba, metal, hip hop e reggae. Para isso, você sabe as características de cada gênero musical e a quantidade de gêneros existentes (nesse exemplo, cinco).

Ao ouvir uma música nova, você analisa seus atributos: Percebe-se o ritmo mais binário, sua progressão de acordes e seus instrumentos mais característicos, como o pandeiro ou o repique, para, no final, caracterizá-la como o gênero Samba!

Esse processo é similar ao que um modelo de classificação treinado realiza. Para treiná-lo, é necessário, como entrada, um conjunto de dados já rotulados (com os gêneros musicais) e o modelo apenas tenta adivinhar qual são as suas classes.

Percebe-se que, como temos que ter um conjunto de dados já rotulados, os modelos de classificação são modelos supervisionados.


### **3.5.2 Clusterização**


Já na clusterização, você recebe um conjunto de músicas sem saber quais são seus gêneros. O seu trabalho é analisar suas características e agrupá-las em gêneros que você criou, juntando as músicas mais parecidas. Por exemplo, sua análise percebeu que existem músicas que possuem um grande destaque ao baixo e à bateria, portanto, decidiu juntá-las em um grande grupo chamado “hip hop-reggae”. Percebe-se que, como utilizamos apenas essa característica, criamos um novo gênero que combina as músicas de hip hop e reggae.

Os modelos de clusterização agem dessa mesma maneira: recebem um grande número de dados e os agrupam em classes baseadas em uma ou mais características e padrões encontrados nesse mesmo conjunto. Diferentemente dos modelos de classificação, não precisam receber os dados já categorizados, são modelos de aprendizado não supervisionados.


### **3.5.3 Classificação x Clusterização**


|             | **Classificação**                                    | **Clusterização**                                                      |
|-------------|--------------------------------------------------|--------------------------------------------------------------------|
| **Objetivo**    |     Classificar dados em classes pré-fixadas     | Agrupar dados com características semelhantes em clusters (grupos) |
| **Propósito**   | Prever a classe de uma instância ainda não vista |         Descobrir padrões e estruturas do conjunto de dados        |
| **Treinamento** |                  Supervisionado                  |                         Não supervisionado                         |
| **Saída**       |              Atribuição à uma classe             |                  Agrupamentos do conjunto de dados                 |

## **2.6 Tipos de tarefas em UL (visão geral)**

**Clustering**: Agrupar dados em clusters (classes) baseados em suas características

**Redução de dimensionalidade**: Reduzir o número de variáveis de um conjunto mantendo, ainda, o máximo de informações relevantes.

É notável que o modelo realiza esse processo ao identificar os padrões e, assim, excluir as variáveis que não possuem muita contribuição para essas estruturas. Isso é feito para simplificar os dados, facilitando a sua visualização, além de possivelmente remover ruídos do conjunto.

**Detecção de outliers**: Pela mesma lógica, é possível utilizar os modelos não supervisionados para encontrar os outliers. Como o modelo já explora os padrões e correlações das variáveis dos dados, basta encontrar as que não fazem parte desta estrutura.

É importante destacar que esse processo não serve apenas para a limpeza de dados, mas, também, para detectar eventos raros que, ao depender do caso de uso, merecem uma atenção maior na análise exploratória.

**Análise exploratória**: Agora que você entende como o modelo age para as tarefas acima, é fácil entender o porquê do modelo ser útil na análise do conjunto de dados.

Utilizando isso como ferramenta, podemos criar visualizações, gráficas e não gráficas para obter um entendimento básico do conjunto de dados de forma clara, simples e autoexplicativa.


## **2.7 Noção de “qualidade” em UL:**

Avaliar os resultados obtidos por aprendizado não supervisionado é bem mais difícil do que com aprendizado supervisionado, pois não temos um gabarito com o qual comparar, portanto não é possível obter métricas como acurácia, recall, precisão, f1-score e outras das medidas mais importantes para aprendizado supervisionado. As opções são bem mais limitadas, geralmente avaliar o resultado envolve retirar novas medidas dos dados após a aplicação do algoritmo, ao invés de comparar com um valor externo ao treino.

A técnica mais fácil de avaliar são os clusters, nesse caso podemos analisar o quão bem nosso algoritmo conseguiu dividir os dados. Quanto mais diferentes os dados em clusters diferentes e quanto mais parecidos os dados em mesmos clusters melhor.

Outras tarefas, como redução de dimensionalidade, são mais difíceis de avaliar. As formas existentes de avaliar de forma objetiva, sem depender de outras técnicas, envolvem olhar para a distribuição dos dados. Queremos saber como o que fizemos alterou métricas como variância, a separação das observações em classes e o quanto conseguimos reter as informações dos dados originais. Por exemplo, uma forma de avaliar algoritmos de redução de dimensionalidade é analisar a relação entre dois pontos estarem distantes um do outro no conjunto de dados original e no conjunto de dados transformado.

Outra maneira de tentar avaliar técnicas de aprendizado não supervisionado é fazer uso de diferentes formas de visualização e outras ferramentas de análise exploratória e tentar “manualmente” compreender se as transformações criadas fazem sentido. Por exemplo, podemos analisar os clusters criados e ver se batem com as expectativas, ou ao tratar de redução de dimensionalidade podemos plotar os dados e ver se surgiram padrões que podemos analisar.

Porém, em muitos casos pode ser difícil avaliar a qualidade dos resultados isoladamente. Uma das melhores formas de checar se a transformação criada foi boa é usá-la em conjunto com outro algoritmo e checar se ela melhora a performance desse algoritmo. Podemos usar redução de dimensionalidade em conjunto com muitos outros algoritmos, incluindo clusterização. Através dessa perspectiva, o uso de técnicas de aprendizado não supervisionado e as diferentes formas de aplicá-las podem ser vistas como novos hiperparâmetros a serem ajustados.


# 3 Fluxo de trabalho da análise exploratória

## **3.1 Definir o objetivo do projeto**

É essencial saber quais perguntas estamos tentando responder ao realizar a análise exploratória. O maior caso exemplo é o esclarecimento e a garantia de que os stakeholders estão realizando as perguntas corretas. O processo de aed pode se tornar muito confuso muito rapidamente e, por isso, deixar explícita a intenção do projeto é umas das formas que podemos evitar esse problema.

É aqui que determinamos o tamanho do escopo da entrega que devemos realizar. Para isso, podemos realizar uma coleta de requisitos via perguntas que podem ser traduzidas em métricas nos conjuntos de dados. Se temos a compreensão clara dos objetivos do projeto, podemos realizar uma análise bem mais efetiva e detalhada dos dados.

Por exemplo, o termo “vendas” pode significar vários atributos: Receita, Lucro, Quantidade, entre outros. Torna-se interessante apurar a métrica, sua dimensão (tempo, região, produto, ..) e classificar o que devemos entregar (para quem iremos apresentar os resultados, qual o contexto que a análise será utilizada, …) para definirmos uma intenção bem clara e direta para a análise que iremos realizar.

## **3.2 Entendendo o contexto dos dados**

Analisar os números brutos é ótimo, mas todo o processo pode ser muito mais fácil e trazer resultados melhores se já tivermos algum conhecimento prévio sobre os dados. É importante nos familiarizarmos com o contexto por trás do problema que estamos analisando, não apenas seguir um passo a passo robótico.

Um exemplo simples disso seria se no meu conjunto de dados eu tivesse 40 features e duas delas fossem “tempo (em meses)” e “tempo (em anos)”, eu não preciso criar uma matriz de correlação de 40x40 e comparar cada par para saber que essas duas features estão altamente correlacionadas, mas nesse caso até um robô chegaria nessa conclusão eventualmente. Porém, um exemplo mais crítico seria, por exemplo, se estivéssemos analisando um conjunto de dados não rotulado com algumas espécies de plantas e quiséssemos tentar criar clusters para separar as espécies, mas sabemos que algumas das espécies podem ter uma variação muito alta de tamanho, então precisamos tratar corretamente dessa medida para que nosso algoritmo não acabe tentando separar as plantas por tamanho e cometa um erro.

## **3.3 Visão geral dos dados**


A análise inicial não deve ser complicada e utilizada para explorar e aprofundar insights dos dados. Seu objetivo é entender o que está acontecendo no conjunto, encontrar quaisquer tendências, padrões e/ou anomalias óbvias e saber se é possível realizar as questões realizadas aos stakeholders anteriormente.

É recomendável que exista uma visualização descritiva geral dos dados antes de qualquer análise mais aprofundada. Assim, torna-se mais fácil as etapas subsequentes e a identificação de qualquer possível erro presente no conjunto de dados.

Antes das análises em si, deve-se identificar os tipos de variáveis presentes: variáveis quantitativas (numéricas) ou variáveis qualitativas (categóricas). Percebe-se que é necessário um reconhecimento da característica de cada tipo de variável, como já discutido anteriormente em 2.1.

Não iremos nos aprofundar muito sobre as métricas que irão ser apresentadas por agora, já que o objetivo deste material é apenas apresentar os conceitos básicos da análise exploratória. Porém, caso queira se aprofundar mais sobre esse assunto, recomenda-se o material “Exploratory Data Analysis - Capítulo 4” apresentado na bibliografia.

## **3.4 Análise Univariada**


### **3.4.1 Análise não gráfica**


Para variáveis categóricas, é possível utilizar apenas a tabela de frequências para a análise, que é a melhor ferramenta não gráfica para a análise exploratória univariada com variáveis categóricas.

Já para variáveis numéricas, temos a inferência estatística como nossa maior aliada. Podemos utilizar ferramentas como a média da amostra e a sua robustez junto com a moda ou mediana (para dados assimétricos) para realizar uma avaliação da tendência central ou realizar uma avaliação da dispersão dos dados pela variância amostral (ANOVA) e seu desvio padrão.

Além disso, também é possível utilizar métricas como o intervalo interquartil, skewness e kurtosis para essa análise.

### **3.4.2 Análise gráfica**

A análise gráfica utiliza-se bem menos de métricas estatísticas, podemos utilizar plots como histogramas, box plots, stem-and-leafs e de quantil-normal. Observe que o único plot utilizado para variáveis qualitativas é o histograma, o restante serve apenas para variáveis quantitativas.

É notável que esses gráficos podem ajudar em qualquer momento da análise exploratória, ao serem utilizados para verificar se a assimetria dos dados, regiões de maior concentração de valores, detecção de outliers, entre outros.

## **3.5 Análise Multivariada**

### **3.5.1 Análise não gráfica**

Para variáveis categóricas, é possível utilizar ferramentas como tabulação cruzada (a melhor para avaliação bivariada, ou seja, entre duas variáveis) e correlação.

Para a análise entre variáveis categóricas e quantitativas, podemos separar os dados em grupos por categorias e analisá-los como se fosse um conjunto univariado. Após isso, podemos comparar cada grupo entre si, o que é, de maneira informal, equivalente à uma análise ANOVA de variância.

Para variáveis quantitativas, podemos calcular a correlação/covariância e suas matrizes.


### **3.5.2 Análise gráfica**


Para a análise entre duas variáveis categóricas, o gráfico mais utilizado é o barplot.

Para uma variável categórica e uma quantitativa, similar à análise não gráfica, podemos plotar um boxplot side-by-side (separando em grupos pelas categorias).
Já para duas variáveis quantitativas, podemos plotar scatterplots ou boxplots para a análise.

## **3.7 Pré-processamento e próximos passos**


A análise exploratória nos permitiu verificar algumas características possíveis dentro do conjunto de dados que podem nos atrapalhar na modelagem de dados. Para isso, podemos aplicar estratégias de pré-processamento no conjunto que estamos trabalhando, o que revisaremos neste tópico.

Podemos avaliar os valores faltantes (missing) e aplicar diversas estratégias de remoção de linhas/colunas ou preenchimento e/ou imputação de valores.Também é importante olhar para a escala dos dados e suas distâncias, o que pode ser sensível às diferenças de valores entre as variáveis. Podemos tratar utilizando estratégias de transformação de escala (padronização e normalização).

Por fim, podemos utilizar encoding para traduzir as variáveis categóricas em uma representação numérica adequada para o modelo não supervisionado.

Assim, após realizar a análise exploratória do conjunto de dados e passar por estratégias de pré-processamento, podemos passar pelo modelo. Temos que os próximos passos são a escolha de qual estratégia não supervisionada vai ser utilizada e sua modelagem.
